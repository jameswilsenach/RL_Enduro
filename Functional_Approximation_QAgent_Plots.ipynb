```python
import numpy as np
import matplotlib.pyplot as plt
rewards = np.load('rewards.npy')
thetas = np.load('thetas.npy')
theta_conv = np.load('theta_conv.npy')
%matplotlib inline

#Plot of rewards vs training episode
fig1 = plt.figure()
plt.plot(np.arange(len(rewards))+1,rewards)
plt.xlabel('episode')
plt.ylabel('reward')
plt.show()
fig1.savefig('rewards.pdf',format='pdf')

#Weights of features by approximate (linear) Q-function
fig2 = plt.figure()
plt.bar(np.arange(len(thetas)),thetas,width=1)
plt.xlabel('index')
plt.xticks(np.arange(len(thetas)+1),fontsize=8)
plt.ylabel('weight')
plt.xlim([0,28])
plt.show()
fig2.savefig('thetas.pdf',format='pdf')

#Rate of change of linear parameters
fig3 = plt.figure()
plt.plot(np.arange(2,len(theta_conv)+2),theta_conv)
plt.xlabel('episode')
plt.ylabel('rate')
plt.show()
fig3.savefig('theta_conv.pdf',format='pdf')
```

```python
fig4 = plt.figure()
plt.hist(rewards)
plt.xlabel('reward')
plt.ylabel('frequency')
plt.show()
fig4.savefig('l_distro.pdf',format='pdf')
```
